## 1. Intro

An artificial neural network --- ANN or just NN, is simply a composition of functions with free 'trainable' parameters, or 'weights'.
In this seminar, we will be talking about NNs that are feedforward, i.e don't have cyclic compositions of the functions,
and we will be talking about them in the context of approximation tasks.

More formally, for any functions $f_i$
$$ f_1(x|w_1),\quad f_2(x, f_1(x)|w_2),\quad f_n(x, f_1(x), \ldots, f_{n-1}(\ldots)|w_n),$$
$F(x) = f_n(\ldots)$ is a feedforward neural network with weights $w_1,\ldots,w_n$,
and we will want it to approximate something, e.g another function.

For instance, let's say that we want to find a linear fit to this:

```python
points_n = 100
x = np.random.rand(points_n) * 10 - 5
y = x * 3 + 4
y += np.random.randn(points_n) / 2
plt.scatter(x, y)
```

We can call function $F(x) = ax + b$ a neural network with weights $a$ and $b$, and train this neural network, which means find such values of $a$ and $b$ that $F(x)$ fits the data above.

NNs are trained with backpropagation algorithm, which is basically
1. Calculate the value of the loss function, that represents how far from desired the output of the network is
2. Calculate the derivative of the loss w.r.t to each trainable parameter of the network
3. Update the values of the parameters with some rule using these derivatives
4. Repeat steps 1-3 for N iterations, or until the loss is low enough, or something else

## 3. Automatic differentiation in PyTorch

PyTorch can automatically calculate derivatives of predefined functions and their compositions.
To indicate that we want to calculate derivative of something w.r.t some tensor, we need to set the `requires_grad` flag of this tensor to True.

```python
x = torch.tensor([1.])
x.requires_grad_()
y = torch.sin(x)
```

To calculate $\partial_x y(x)$ we need to call `y.backward()` (notice, not `y.backward(x)`). The value of the derivative is then stored in `x.grad`.

```python
y.backward()
print(x.grad, '==', torch.cos(x))
```

Same for composition of functions.

```python
x = torch.tensor([2.], requires_grad=True)
y = x ** 2 + 1
y.backward()
print(x.grad, '==', 2 * x)
```

Same for multivariate functions, e.g $y = \sum_i x_i a_i$

```python
x = torch.ones(10, requires_grad=True)
a = torch.arange(len(x), dtype=x.dtype)
y = torch.sum(x * a)
y.backward()

print(x.grad)
print(a)
```

If a tensor already has `grad` attribute, e.g from the previous call to `backward`, subsequent calls will add to the value of this attribute.
So, if you, e.g, `backward()` in a loop, you need to explicitely set `grad` to zero each time.

```python
x = torch.tensor([1.], requires_grad=True)
x.grad is None
```

```python
a = x * 2
b = x * 3
y = a + b  # == x * 5
# in this case y.backward() is the same as
a.backward()
b.backward()
x.grad
```

```python
x = torch.tensor([1.], requires_grad=True)
print(x.grad is None)
a = x * 2
b = x * 3
a.backward()
print(x.grad)
b.backward()
print(x.grad)
```

```python
x = torch.tensor([1.], requires_grad=True)
for i in range(2):
    y = x * 2
    y.backward()
    print(x.grad)
```

```python
x = torch.tensor([1.], requires_grad=True)
for i in range(2):
    y = x * 2
    y.backward()
    print(x.grad)
    x.grad.zero_()
```

Freezing layers is a technique where we prevent certain layers from updating their weights during training. This can be useful in situations where the features learned by those layers are already good enough and don’t need to be adjusted. It’s especially common in transfer learning, where you might freeze the early layers of a pre-trained model and only fine-tune the later layers for a new task.

### Weight Initialization

 If the weights are initialized poorly, it can lead to issues like vanishing or exploding gradients. Proper initialization can help us improve the convergence speed and stability of the training process.

1. **Xavier Initialization (Glorot Initialization)**:
  This method aims to keep the variance of the activations roughly the same across layers. It’s typically used for networks with sigmoid or tanh activations, which can saturate and cause gradient issues if not initialized correctly.
  The weights are drawn from a distribution with variance `2 / (n_in + n_out)`, where `n_in` and `n_out` are the number of input and output neurons.

2. **Kaiming Initialization (He Initialization)**:
  This initialization is specifically designed for networks with ReLU activations. Since ReLU doesn’t saturate but has its own issues with exploding gradients, Kaiming initialization helps mitigate these problems.
  Weights are drawn from a distribution with variance `2 / n_in`, where `n_in` is the number of input neurons.


```python
hidden_dim_size = 80
input_size = 8 * 8
number_of_classes = 10

# Define two networks: one for proper (good) initialization and one for bad initialization
net_good = torch.nn.Sequential(
    torch.nn.Linear(input_size, hidden_dim_size, bias=True),
    torch.nn.ReLU(),
    torch.nn.Linear(hidden_dim_size, number_of_classes, bias=True)
)
net_bad = torch.nn.Sequential(
    torch.nn.Linear(input_size, hidden_dim_size, bias=True),
    torch.nn.ReLU(),
    torch.nn.Linear(hidden_dim_size, number_of_classes, bias=True)
)
```

Proper weight initialization helps to keep the variance of activations under control,
ensuring a stable training process. Here, we use Kaiming Uniform initialization (suitable for ReLU)
